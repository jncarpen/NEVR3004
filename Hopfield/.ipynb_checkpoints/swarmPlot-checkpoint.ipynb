```python
# import libraries

import scipy.io as sio
import pandas as pd
import numpy as np
from array import *
import scipy.stats as stats
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# import matfile as dictionary
path = "C:\\Users\\17145\\Documents\\code\\NEVR3004\\Hopfield\\data\\full2.mat" #NS0
matfile = sio.loadmat(path) # load .mat file

# pull out cell arrayb
minCon = matfile['minCon']
Convergence = matfile['Convergence']
Patterns = matfile['allPatterns']

# convert numpy array to df
NS0 = pd.DataFrame(Convergence[0][5][:,1:10:2], columns = ['10%', '30%', '50%', '70%', '90%']) # pull *every other* noise category
NS0.head()

P = pd.DataFrame(np.transpose(Patterns))
# P.head()
```

```python
# Compute the correlation matrix
corr = P.corr()

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(18, 18))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(220, 10, as_cmap=True)

# # Draw the heatmap with the mask and correct aspect ratio
# sns.heatmap(corr, cmap=cmap, vmax=.3, center=0,
#             square=True, linewidths=.5, cbar_kws={"shrink": .5})

sns.heatmap(corr, vmin=-1, vmax=+1, center=0, annot=True, square=True, cbar_kws={"shrink": .82});
plt.savefig('correlationMat.png')
```

```python
# import matfile as dictionary
path = "C:\\Users\\17145\\Documents\\code\\NEVR3004\\Hopfield\\data\\HopDataset_NS2.mat"
matfile = sio.loadmat(path) # load .mat file

# pull out cell arrayb
minCon = matfile['minCon']
Convergence = matfile['Convergence']

# convert numpy array to df
NS0 = pd.DataFrame(minCon[0][1][:,1:10:2], columns = ['10%', '30%', '50%', '70%', '90%']) # pull *every other* noise category
NS0.head() # display
```

```python
# generate categorical scatterplot with non-overlapping points (S0)

a4_dims = (14, 8) # scale figure
sns.set(font_scale = 2, style = 'white', palette='bone')
swarm = sns.swarmplot(data=S0, size = 1.5, edgecolor = 'k');
sns.despine(top=True, right=True)
axes = swarm.axes # grab axes object
axes.set_ylim(0,500); # match axes
plt.savefig('swarmS0_7P.png')
```
