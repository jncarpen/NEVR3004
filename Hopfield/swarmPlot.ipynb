```python
# import matfile as dictionary
path = "C:\\Users\\17145\\Documents\\code\\NEVR3004\\Hopfield\\data\\HopDataset_S0.mat"
matfile = sio.loadmat(path) # load .mat file

# pull out cell arrayb
minCon = matfile['minCon']
Convergence = matfile['Convergence']

# convert numpy array to df
S0 = pd.DataFrame(minCon[0][0][:,1:10:2], columns = ['10%', '30%', '50%', '70%', '90%']) # pull *every other* noise category
S0.head() # display
```
